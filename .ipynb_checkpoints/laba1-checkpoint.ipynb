##### **Задача 1.** На данных своего варианта повторить расчёты и построить графики из первой лабораторной. Пояснить выбор наилучшего количества степеней свободы.

```python
 # загрузка пакетов
import numpy as np
import random
```

```python
np.random.seed(22)

# всего наблюдений
n_all = 60

# доля обучающей выборки
train_percent = 0.85

# параметры распределения X
x_min, x_max = 5, 105

# генерируем X
x = np.random.uniform(x_min, x_max, n_all)

# генерируем случайный шум
mu, sigma = 0, 1
res = np.random.normal(mu, sigma, n_all)

# отбираем наблюдения в обучающую выборку
in_train = np.random.choice(n_all, int(n_all*train_percent), replace=False)

# истинная функция взаимосвязи
def y_func (x):
    return  17 + 2e-02*x - 5e-03*(x-45)**2 + 6e-05*(x-54)**3
    
# для графика истинной взаимосвязи
x_line = np.linspace(x_min, x_max, n_all)
y_line = y_func(x_line)

# фактические значения y (с шумом)
y = y_func(x) + res

# создаём векторы с данными для построения графиков наблюдения на обучающей выборке
x_train = x[in_train]
y_train = y[in_train]

# наблюдения на тестовой выборке
x_test = np.delete(x, in_train)
y_test = np.delete(y, in_train)
```

```python
# исходные данные на графике

import matplotlib.pyplot as plt

plt.scatter(x_train, y_train, c='black', label='обучение', alpha=0.6)
plt.scatter(x_test, y_test, c='red', label='тест', alpha=0.6)
plt.plot(x_line, y_line, c='black', linestyle='dashed', label='f(X) = 17+0.02x−0.005(x−45)^3+0.00006(x−54)^3')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# загрузка пакетов
from pandas import DataFrame
import rpy2.robjects as robjects
```

```python
# максимальное число степеней свободы для модели сплайна
max_df = 40

# создаём из данных объекты, понятные R
r_x = robjects.FloatVector(x_train)
r_y = robjects.FloatVector(y_train)

# импортируем нужную функцию из R
r_smooth_spline = robjects.r['smooth.spline']

# фрейм данных для сохранения MSE на обучающей и на тестовой при разных степенях свободы
MSE_all_dfs = [[0.] * 2 for i in range(2, max_df+1)]
MSE_all_dfs = DataFrame(MSE_all_dfs, columns=['MSE_train', 'MSE_test'],
index=list(range(2, max_df+1)))

# заполняем фрейм
for i_df in range(2, max_df+1) :
    # обучаем модель
    fit = r_smooth_spline(x=r_x, y=r_y, df=i_df)
    
    # прогноз на обучающей выборке
    y_train_hat = np.array(robjects.r['predict'](fit,
        robjects.FloatVector(x_train)).rx2('y'))
    
    # прогноз на тестовой выборке
    y_test_hat = np.array(robjects.r['predict'](fit,
        robjects.FloatVector(x_test)).rx2('y'))
    
    # считаем MSE
    MSE_all_dfs.loc[i_df, 'MSE_train'] = \
        sum((y_train - y_train_hat)**2) / len(y_train_hat)
    MSE_all_dfs.loc[i_df, 'MSE_test'] = \
        sum((y_test - y_test_hat)**2) / len(y_test_hat)
```

```python
MSE_all_dfs[MSE_all_dfs['MSE_test'] == min(MSE_all_dfs['MSE_test'])]
```

```python
plt.plot(MSE_all_dfs.index, MSE_all_dfs['MSE_train'], color='dimgrey', marker='', label='обучение')
plt.plot(MSE_all_dfs.index, MSE_all_dfs['MSE_test'], color='red', marker='', label='тест')
plt.plot([2, 2], [MSE_all_dfs.iloc[0]['MSE_train'], MSE_all_dfs.iloc[0]['MSE_test']], ls='', marker='s', label='df=2')
plt.plot([15, 15], [MSE_all_dfs.iloc[13]['MSE_train'], MSE_all_dfs.iloc[13]['MSE_test']], ls='', marker='s', label='df=15')
plt.plot([38, 38], [MSE_all_dfs.iloc[36]['MSE_train'], MSE_all_dfs.iloc[36]['MSE_test']], ls='', marker='s', label='df=38')
plt.hlines(1, 0, 40, linestyles='dashed', color='grey', label='Var(e)=1')
# подписи осей
plt.xlabel('степени свободы (df)')
plt.ylabel('MSE')
# легенда
plt.legend(loc='best')
plt.show()
```

Оптимальная степень свободы – df=15, так как среднеквадратичная ошибка на обучающей выборке при такой модели минимальная.

##### **Задача 2.** Повторить расчёты, меняя характеристики (долю обучающей выборки) согласно своему варианту. Проанализировать, как меняется MSE при изменении характеристик.

```python
for train_percent in [0.8, 0.85, 0.9]:
    np.random.seed(22)
    x = np.random.uniform(x_min, x_max, n_all)
    res = np.random.normal(mu, sigma, n_all)
    y = y_func(x) + res
    in_train = np.random.choice(np.arange(n_all), int(n_all*train_percent), replace=False)
    x_train = x[in_train]
    y_train = y[in_train]
    x_test = np.delete(x, in_train)
    y_test = np.delete(y, in_train)
    
    r_x = robjects.FloatVector(x_train)
    r_y = robjects.FloatVector(y_train)

    # находим df, оптимальное для данного train_percent
    MSE_all_dfs = [[0.] * 2 for i in range(2, max_df+1)]
    MSE_all_dfs = DataFrame(MSE_all_dfs, columns=['MSE_train', 'MSE_test'], index=list(range(2, max_df+1)))
    
    for i_df in range(2, max_df+1) :
        fit = r_smooth_spline(x=r_x, y=r_y, df=i_df)
        y_hat_train = np.array(robjects.r['predict'](fit, robjects.FloatVector(x_train)).rx2('y'))
        y_hat_test = np.array(robjects.r['predict'](fit, robjects.FloatVector(x_test)).rx2('y'))
        MSE_all_dfs.loc[i_df, 'MSE_train'] = \
            sum((y_train - y_hat_train)**2) / len(y_hat_train)
        MSE_all_dfs.loc[i_df, 'MSE_test'] = \
            sum((y_test - y_hat_test)**2) / len(y_hat_train)

    print(f'train_percent = {train_percent}')
    print(f'df = {MSE_all_dfs[MSE_all_dfs['MSE_test'] == min(MSE_all_dfs['MSE_test'])].index[0]}')
    print(f'MSE = {min(MSE_all_dfs['MSE_test'])}')
    print()
```

**Вывод:** чем больше доля обучающей выборки, тем меньше требуется степеней свободы, и тем меньше становится среднеквадратичная ошибка, то есть обучение становится точнее.
